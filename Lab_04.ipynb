# 1.The USCensus data

```python
import pandas as pd
# The data is wrapped with this notebook.
# This is a smaller version of the 1990 census, available at UCI and other places.
# Taking a subset helps running time and to see the dendrogram.
df = pd.read_csv('Downloads/CS663-hierarchical-data-master/USCensus1990-2000rows.data.txt')
```

```python
data = df.iloc[:100,1:]
data
```

```python
from scipy.cluster.hierarchy import linkage, dendrogram

Z = linkage(df, 'ward')
```

```python
# Draw the dendrogram
import matplotlib.pyplot as plt

%matplotlib inline

plt.figure()
dendrogram(Z)
plt.show()
```

From the above dendrograph we can see that when the height reaches above 15000, data points can be clustered into two groups and whne the height is above 8000, datapoints can be clustered into four groups. Also, as the height decreases, more clusteres are formed.

# 2.NIPS_data

```python
#read data from the csv file
nips_df = pd.read_csv('Downloads/NIPS_1987-2015.csv')
nips_df
```

```python
#extracting the non string data
nips_data = nips_df.iloc[:, 1:]
nips_data
```

```python
# Draw the dendrogram
from scipy.cluster.hierarchy import linkage, dendrogram
import matplotlib.pyplot as plt
%matplotlib inline

#X is the clustered data
X = linkage(nips_data, 'ward')

plt.figure(figsize = (20,20))
dendrogram(X)
plt.show()
```

From the above dendrograph we can see that when the height reaches above 3000, data points can be clustered into two groups and whne the height is above 2000, datapoints can be clustered into four groups. Also, as the height decreases, more clusteres are formed.
